```python
import pandas as pd
import nltk.tokenize.punkt
import pickle
import random
```

```python
df = pd.read_csv('./data/train_task1_latest.csv')
%time russian_text = "\n ".join(random.sample(list(df['paragraph']),len(df)//2))
```

```python
# import punkt 
import nltk.tokenize.punkt 

# Make a new Tokenizer 
tokenizer = nltk.tokenize.punkt.PunktSentenceTokenizer() 

# Train tokenizer 
%time tokenizer.train(russian_text) 

# Dump pickled tokenizer  
out = open("/users/sermolov/nltk_data/tokenizers/punkt/PY3/russian.pickle","wb") 
pickle.dump(tokenizer, out) 
out.close()
```

```python
texts = [
    'В т.н. физике существуют области, в т. ч. Бла-бла бла. В.И.Ленин (XV-XVII вв. н.э )',
    'Реки и озера С.Ш.А. Алфавитный справочник.',
    'Пример, когда и т.д. и т.п. пишутся без пробелов.',
    'Пример, когда и т. д. и т. п. пишутся с пробелами.',
    'Пример, когда предложение заканчиваетс на и т.д. А после идет следующее предложение.',
    'Пример, в котором предложение заканчивается на многоточие ... А потом начинается новое предложение.',
    'Как парсятся другие знаки препинания? Например, восклицательный знак! А восклицательный вопрос!?',
    'Что, если встретятся преносы \n Будут ли они что-то значить? Или \n нет ...',
    'Пример нового предложения без отступа.Вот такое.',
    'Странные знаки перед точкой[6]. Что будет {}//? Посмотрим',
    random.choice(russian_text.split('\n ')),
    random.choice(russian_text.split('\n ')),
    random.choice(russian_text.split('\n '))
]

for txt in texts:
    print('=================')
    for sent in nltk.tokenize.sent_tokenize(
        language='russian', # punkt/PY3/russian.pickle
        text=txt):
        print(sent,end='</>\n')
```
